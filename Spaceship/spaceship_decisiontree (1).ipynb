# Spaceship Decision Tree (DecisionTreeClassifier) Submission

```
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import sklearn 
```

# Bring the Data In

```
X_df = pd.read_csv('spaceship_train_X_v2.csv')
y_df = pd.read_csv('spaceship_train_y.csv')
X_submission = pd.read_csv('spaceship_test_X_v2.csv')

y_df.drop('Unnamed: 0', axis=1, inplace=True)
X_df.head()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, stratify=y_df, test_size=0.2, train_size=0.8, random_state=1)

# Ini buat ganti y_train and y_test ke 1D array
y_train = y_train['Transported'].values
y_test = y_test['Transported'].values
```

# Model!

We'll be playing with the 'MaxDepth' coefficient. The deeper the tree, the model will try to fit the training data more, which idk let's see.

```
from sklearn.tree import DecisionTreeClassifier

# Testing with multiple max-depth
max_depth = range(1, 10)
training_accuracy = []
testing_accuracy = []

for depth in max_depth:
    tree = DecisionTreeClassifier(max_depth=depth, random_state=0).fit(X_train, y_train)
    training_accuracy.append(tree.score(X_train, y_train))
    testing_accuracy.append(tree.score(X_test, y_test))
```

```
tree_performance = {'max_depth':max_depth, 'training_accuracy':training_accuracy, 'testing_accuracy':testing_accuracy}
tree_performance = pd.DataFrame(tree_performance)
sns.lineplot(data=tree_performance, x='max_depth', y='training_accuracy', label='Training accuracy')
sns.lineplot(data=tree_performance, x='max_depth', y='testing_accuracy', label='Testing accuracy')

plt.title("DecisionTreeClassifier Model Performance")
plt.xlabel("Max-Depth")
plt.ylabel("Accuracy")

plt.legend()
plt.show()
```

```
tree_performance.sort_values(by='testing_accuracy', ascending=False)
```

## Takeaway

In this case, we will proceed with max_depth of 7 (previously we've tried with 3 and 6)

# Submission Prediction

```
clf = DecisionTreeClassifier(random_state=0, max_depth=7)

# For training, we use ALL data from spaceship_train_X_v2.csv and spaceship_train_y.csv
clf.fit(X_df, y_df['Transported'].values)

# Generate prediction for spaceship_test_X_v2.csv
y_prediction = clf.predict(X_submission)
print(y_prediction)
```

```
submission_dict = {'PassengerId':X_submission['PassengerId'], 'Transported':y_prediction.astype('bool')}
submission_dict = pd.DataFrame(submission_dict)
submission_dict
```

## Export CSV

```
submission_dict.to_csv('spaceship_decisiontree_md7_submission.csv', index=False)
```
