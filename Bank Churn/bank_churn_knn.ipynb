```
df = pd.read_csv('bankchurn_train.csv', index_col=0)
X_submission = pd.read_csv('bankchurn_test.csv', index_col=0)

df.head()
```

```
dummies_column = ['Geography', 'Gender']
dummies_geogender= pd.get_dummies(df[dummies_column]).astype('int')
dummies_geogender.head()
```

```
dummies_geogender= pd.get_dummies(X_submission[dummies_column]).astype('int')
dummies_geogender.head()
```

```
df = pd.concat([df, dummies_geogender], axis=1)
df.drop(columns=['Geography',"Gender"], inplace=True)
df.head()
```

```
X_submission = pd.concat([X_submission, dummies_geogender], axis=1)
X_submission.drop(columns=['Geography',"Gender"], inplace=True)
X_submission.head()
```

```
from sklearn.preprocessing import StandardScaler

# Pisahkan fitur dan label
X = df.drop(columns='Exited')
y = df['Exited']

# Inisialisasi pipeline untuk Z-score scaling
zscore_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Daftar kolom yang akan di-Z-score
columns_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 
                    'IsActiveMember', 'EstimatedSalary']

# Lakukan Z-score scaling pada fitur
X_zscore_scaled = X.copy()
X_zscore_scaled[columns_to_scale] = zscore_pipeline.fit_transform(X[columns_to_scale])

# Lakukan Z-score scaling pada data pengujian juga
X_submission_zscore_scaled = X_submission.copy()
X_submission_zscore_scaled[columns_to_scale] = zscore_pipeline.transform(X_submission[columns_to_scale])

# Hasil dari Z-score scaling
print("Data pelatihan setelah Z-score scaling:")
print(X_zscore_scaled.head())

print("\nData pengujian setelah Z-score scaling:")
print(X_submission_zscore_scaled.head())
```

```
from sklearn.preprocessing import MinMaxScaler

# Combine training and test data for one-hot encoding
combined_data = pd.concat([X, X_submission])

# Perform one-hot encoding
combined_data = pd.get_dummies(combined_data)

# Split back to training and test sets
X_encoded = combined_data.iloc[:X.shape[0]]
X_submission_encoded = combined_data.iloc[X.shape[0]:]

# Initialize Min-Max scaler
minmax_scaler = MinMaxScaler()

# Fit and transform training data
X_minmax_scaled = minmax_scaler.fit_transform(X_encoded)

# Transform test data
X_submission_minmax_scaled = minmax_scaler.transform(X_submission_encoded)

# Convert back to DataFrame
X_minmax_scaled = pd.DataFrame(X_minmax_scaled, columns=X_encoded.columns, index=X_encoded.index)
X_submission_minmax_scaled = pd.DataFrame(X_submission_minmax_scaled, columns=X_submission_encoded.columns, index=X_submission_encoded.index)

# Display the results
print("Data pelatihan setelah Min-Max scaling:")
print(X_minmax_scaled.head())

print("\nData pengujian setelah Min-Max scaling:")
print(X_submission_minmax_scaled.head())

```
